# Tic Tac Toe Game

This project involves building a Tic Tac Toe game using Python without a graphical user interface(GUI) on 3*3 grid.


```python
def Print_Board(board):
    '''
    print Board of the game with updted moves.
    
    '''
#     Board Design 
    for i,row  in enumerate (board): 
        print(' | '.join(row))
        if i !=2 :
            print('-'*8)


def Make_Moves(board, player):
    '''
    ask player to take right place in board and check if the place taken or not,
    input : list(board), player's symbol
    '''
    while True:
        try:
            row, col = map(int, input('enter row and column (0-2) with separate space :').split())
        except : continue
            
        if (row <0 or row >2 or col < 0 or  col >2 or board[row][col]!=''):
            
            print('you must play in the right place, Try again ')
        else :
            board[row][col] = player
            break
            
def Check_Win(board, player):
    '''
    check whose player win, 
    the player win if rows, columns, or diagonals of the board contain three of the player's symbols
    input : list(board), player's symbol
    output : bool / True if one of the game rule acheived 
    '''
    rows = []; cols = []
    left_diagonal = []; right_diagonal= []

    for col in range(len(board)): # itertate in rows 
        for row in range(3): # iterate in columns
            rows.append(board[row][col]) # appending rows
            cols.append(board[col][row]) # appending columns

            if col == row:
                left_diagonal.append(board[row][col]) #appending left_diagonal 
                
            if (row + col)  == len(board)-1:
                right_diagonal.append(board[row][col]) # appending right_diagnoal
                
                
         # check if any column or row is full with currern player's symbol       
        if rows.count(player) == len(board) or cols.count(player) == len(board): 
            return True
        else:
            rows = [];cols = []
    # check if left diagonal or right diagonal is fulled with current player's symbol
    if left_diagonal.count(player) == len(board) or right_diagonal.count(player)==len(board):
        return True
    
    return False

def Check_Tie(board):
    '''
    check draw, draw happen if board full with player'symbol and no one of player win
    input : list(board) 
    output : bool / True if threre 
    '''
    for i in board :
        for j in i :
            if j =='': #check if any place is empty
                return True 
    return False

def main(): # main function
    '''
    main function where the game start
    '''

    board = [['','',''],['','',''],['','','']] # board of the game 3*3

    current_player = 'X'
    while True:   # Main Game Loop
        Print_Board(board) 
        print (f'player {current_player} Turn')

        Make_Moves(board,current_player) 
        win = Check_Win(board, current_player)
        if win == True : # check if current player win 
            Print_Board(board)
            print (f'Player {current_player} Wins!')
            break
        tie = Check_Tie(board)
        if tie == False : # chek if the game ends with draw
            Print_Board(board)
            print('Tie!')
            break
            
        # take turns   
        if current_player == 'X':
            current_player = 'O'
        else :
            current_player = 'X'
if __name__ == '__main__':
    main()
```
